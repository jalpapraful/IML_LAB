```python
#Gadhiya Jalpa
#216270307103
#Write a Pandas program to implement following operationto convert a NumPy array to a Pandas series

a=np.array([1,2,3,4,5])

b=pd.Series(a)
print("pandas series :",b)
```

```python
#Gadhiya Jalpa
#216270307103
#Write a Pandas program to implement following operation to convert the first column of a DataFrame as a Series
data=pd.DataFrame({'colors': ['white','black','pink','blue','red'],'number':[1,2,3,4,5]})
print("Dataframe is:",data)

series=data.iloc[:,0]
print("convert the first column of a DataFrame as a Series:\n",series)
```

```python
#Gadhiya Jalpa
#216270307103
#Write a Pandas program to implement following operation to create the mean and standard deviation of the data of a given Series
import numpy as np 
import pandas as pd

series1=pd.Series([2,4,6,8,10])
print("Pandas series:",series1)

smean=series1.mean()
print("\n mean is:",smean)

sestd=series1.std()
print("\n standard deviation is:",sestd)
```

```python
#Gadhiya Jalpa
#216270307103
#Write a Pandas program to implement following operation to sort a given Series
series1=pd.Series([7,10,3,4,9,2,10,1])
print("orignal series is:",series1)

series_sort=series1.sort_values()
print("\n sorting series is:",series_sort)
```
